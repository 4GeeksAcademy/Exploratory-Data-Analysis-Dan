```python
import pandas as pd

ab_data = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/data-preprocessing-project-tutorial/main/AB_NYC_2019.csv")
ab_data
```

```python
ab_data.columns
```

```python
ab_data = ab_data.drop(["host_id", "host_name", "id"], axis = 1)
```

```python
ab_data.shape
ab_data.info()
```

```python
ab_data.duplicated().sum()
ab_data.info()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.histplot(ab_data, x = "neighbourhood_group")
plt.show()
```

```python
cheaper_than_200 = ab_data["price"] < 200
ab_data[cheaper_than_200]
sns.histplot(ab_data[cheaper_than_200]["price"], bins=10)
```

```python
less_than_14_days = ab_data["minimum_nights"] < 14
ab_data[less_than_14_days]
sns.histplot(ab_data[less_than_14_days]["minimum_nights"], bins=14)
plt.show()
```

```python
fig, axis = plt.subplots(1, 3, figsize = (10, 8))

sns.histplot(ax = axis[0], data = ab_data, x = "neighbourhood_group", color="red")
axis[0].tick_params(axis='x', rotation=90)
sns.histplot(ax = axis[1], data = ab_data[cheaper_than_200]["price"], bins=10, palette="inferno", color = "green")
sns.histplot(ax = axis[2], data = ab_data[less_than_14_days]["minimum_nights"], bins=14, color = "purple")
plt.tight_layout()
plt.show()
```
